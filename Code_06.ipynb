**Q1: In your own number system (i.e., in base 15), represent your number as a list of digits and create a function to add 1 to your number.**

```python
def digitvalinbase10(element):
    syms = "0123456789ABCDE"
    for i, symbol in enumerate(syms): #use enumerate to get the index.
        if symbol == element:
            return i
    return -1
```

```python
def add1tonumberinbase15(numberarray):
    reverse = numberarray[::-1]
    base15digits = "0123456789ABCDE"
    add = 1
    position = 0 #starting from left 
    result = []
    while position <= len(reverse)-1 or add:
        if position <= len(reverse)-1:
            digitvalueinbase10 = digitvalinbase10(reverse[position])  
            position += 1
        else:
            digitvalueinbase10 = 0

        newvalueinbase10 = digitvalueinbase10 + add

        if newvalueinbase10 >= 15:
            add = 1
            result.append("0")
        else:
            add = 0
            result.append(base15digits[newvalueinbase10])

    return result[::-1]   
```

```python
add1tonumberinbase15(["1", "2", "E"])
```

```python
add1tonumberinbase15(["D", "E", "E"])
```

```python
add1tonumberinbase15(["9"])
```

```python
add1tonumberinbase15(["9","5","E"])
```

```python
add1tonumberinbase15(["E","E","E"])
```

**Q2: Print all numbers of 4 digits in base 15.
    Start with [0,0,0,0] and keep adding 1  to it using previously coded add1tonumberinbase15() until the list exhausts.**

```python
def all4digitnumberinbase15():
    startingnum = ['0','0','0','0']
    maxnum = ['E','E','E','E']
    while True:
        print(startingnum)
        if startingnum == maxnum:
            break
        else:
            startingnum = add1tonumberinbase15(startingnum)  
```
